```python
# Import 
import cv2
import os

# Lokasi Folder Dataset
pathDataset = r"C:\Users\Asus\Desktop\python\komvis\Dataset"

# Membuat Inputan Untuk Nama Dataset
name = input("Masukan Nama Anda : ")

# Membuat Fungsi Create Folder Otomatis
def createFolder():
    if os.path.isdir(pathDataset): # Memastikan Ada pathDataset
        os.chdir(pathDataset) # Ubah Directory Saat Ini Ke Directory Yang Ditentukan
        os.mkdir(name) # Membuat Folder Dataset Sesuai Nama Yang Dimasukan
        
# Membuat Fungsi Split Video Ke Gambar
def splitingVideo(fileNameVideo):
    video = cv2.VideoCapture(fileNameVideo)
    count = 1
    while(video.isOpened()):
        ret,frame = video.read()
        if ret == False:
            break
        # Membuat File Name Gambar Hasil Split Yang Disimpan    
        fileNameGambar = name + "-" + str(count) + '.jpg'
        # Simpan Gambar
        cv2.imwrite(fileNameGambar,frame)
        print(fileNameGambar, "Berhasil Tersimpan!")
        count += 1
        
# Memanggil Fungsi
createFolder()  

# Membuat File Name Video
fileName = name + '.avi'

# Membuka Kamera
camera = cv2.VideoCapture(0)

# Ambil Ukuran Tinggi Lebar Kamera
frame_lebar = int(camera.get(3))
frame_tinggi = int(camera.get(4))

# Kode Compress Frame
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# Menggabungkan Folder Dataset Dengan Folder Nama Untuk Menyimpan Video
outputGabung = os.path.join(pathDataset,name+"\\")
os.chdir(outputGabung)
outputVideo = cv2.VideoWriter(fileName,fourcc,20,(frame_lebar,frame_tinggi)) # Koding Untuk Output Video

if camera.isOpened() == False:
    print("Gagal Membuka Kamera!")
while(camera.isOpened()):
    ret,frame = camera.read() 
    
    # Simpan Video Frame Per Frame
    outputVideo.write(frame)
    cv2.imshow("Video",frame)    
    
        
camera.release()
cv2.destroyAllWindows()

# Memanggil Fungsi Spliting Video
splitingVideo(fileName)
```
